```python
import numpy as np
import matplotlib.pyplot as plt

# Edades y síntomas
edades = [30, 40, 60]
sintomas = [1, 2, 3]  # 1: dolor de cabeza, 2: migraña y vértigo, 3: parálisis de la pierna derecha

# Rango de edades
rango_edades = np.arange(30, 61)

# Interpolación para obtener los valores de síntomas en cada edad
sintomas_interp = np.interp(rango_edades, edades, sintomas)

# Cálculo de la derivada
derivada_sintomas = np.gradient(sintomas_interp, rango_edades)

# Gráfico de la derivada de los síntomas
plt.plot(rango_edades, derivada_sintomas)
plt.xlabel("Edad")
plt.ylabel("Derivada de los Síntomas")
plt.title("Derivada de los Síntomas de una Persona a lo largo del Tiempo")
plt.xticks(np.arange(30, 61, 5))
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
